```python
# importing libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing csv-data file 
customers = pd.read_csv("C:/Users/Loretta/Desktop/SkillHarvest Assignment/Customers.csv")
customers
```

<h3>Data Cleaning & Wrangling</h3>

```python
#shape(): To displays the total number of rows and columns.
customers.shape
```

```python
# gives you information about your customers
customers.info()
```

```python
customers.columns
```

```python
donts = ['Prefix', 'FirstName', 'LastName', 'Unnamed: 17', 'Unnamed: 18', 'BirthDate.1', 'Unnamed: 8', 'Unnamed: 5', 'Unnamed: 6']
```

```python
#to remove all columns in the 'dont' list
customers.drop(donts, axis = 1, inplace = True)
```

```python
customers.shape
```

```python
 #sample():to displays a sample of rows of all customers, randomly selected
customers.sample(8)
```

```python
#to see the columns that has missing items
customers.isna().sum()
```

```python
#forward fill uses what's at the bottom to fill what's at the top
customers.fillna(method = 'ffill', inplace = True)
```

```python
customers.isna().sum()
```

```python
#to rename the 4th column
customers.rename(columns = {'Unnamed: 4':'Full Name'}, inplace = True)
```

```python
customers.sample(4)
```

```python
#str.strip() is used to remove a particular xter totally without replacing it with anything; $ sign in AnnualIncome make it a string, so we'd remove it.
customers.AnnualIncome = customers.AnnualIncome.str.strip('$')
customers.AnnualIncome
```

```python
#str.replace() is used to replace a particular character with another character or nothing.
customers.AnnualIncome = customers.AnnualIncome.str.replace(',', '')
customers.AnnualIncome
```

```python
#astype is for type casting, used to convert one datatype of a column to another; changing our currency to an integer
customers.AnnualIncome = customers.AnnualIncome.astype(int)
```

```python
#to change the datatype of the date from object(string) to date datatype; %d - day, %m - month, %Y - year 
customers.BirthDate = pd.to_datetime(customers.BirthDate, format = '%d/%m/%Y')
```

```python
customers.sample(4)
```

```python
customers.info()
```

```python
#to replace some characters with their full name
customers.MaritalStatus = customers.MaritalStatus.replace('M', 'Married')
customers.MaritalStatus = customers.MaritalStatus.replace('S', 'Single')
customers.Gender = customers.Gender.replace('M', 'Male')
customers.Gender = customers.Gender.replace('F', 'Female')
customers.HomeOwner = customers.HomeOwner.replace('Y', 'Yes')
customers.HomeOwner = customers.HomeOwner.replace('N', 'No')
```

```python
customers.sample(6)
```
